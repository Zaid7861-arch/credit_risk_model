```python
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, accuracy_score

# 1. Generate simulated Customer Credit Data (Income, Debt, Payment History, etc.)
X, y = make_classification(n_samples=5000, n_features=8, random_state=100)
df = pd.DataFrame(X, columns=['Income', 'Debt_Ratio', 'Late_Payments', 'Credit_History', 'Age', 'Employment_Length', 'F2', 'F3'])
df['Will_Default'] = y

# 2. Scale the data (Standardization is crucial for Logistic Regression)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.drop('Will_Default', axis=1))

# 3. Train / Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, df['Will_Default'], test_size=0.2, random_state=100)

# 4. Train Logistic Regression Model
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# 5. Predict and Evaluate
preds = lr_model.predict(X_test)
probs = lr_model.predict_proba(X_test)[:, 1]

print("--- Credit Risk Underwriting Results ---")
print(f"Accuracy: {accuracy_score(y_test, preds):.2%}")
print(f"AUC-ROC Score: {roc_auc_score(y_test, probs):.3f} (Measures model ability to distinguish default vs non-default)")

```
